```python
#Create a copy and inspect
movies_cleaned = movies_df.copy()
movies_cleaned.info()
movies_cleaned.head()
```

```python
# Summary of missing values
missing = movies_cleaned.isnull().sum().sort_values(ascending=False)
missing_percent = (movies_cleaned.isnull().sum()/len(movies_cleaned)*100).sort_values(ascending=False)
missing_df = pd.concat([missing, missing_percent], axis=1, keys=['Missing Values', 'Percent'])
missing_df
```

```python
# Drop all rows with any null values
movies_cleaned = movies_cleaned.dropna()

# Quick check to confirm
print(f"Shape after dropping rows with nulls: {movies_cleaned.shape}")
print(movies_cleaned.isnull().sum())
```

```python
movies_cleaned.describe()
```

```python
# List of columns to drop
cols_to_drop = ['overview', 'keywords', 'tagline', 'poster_path', 'backdrop_path', 'recommendations']

# Drop columns
movies_cleaned = movies_cleaned.drop(columns=cols_to_drop)

movies_cleaned.info()
movies_cleaned.head(20)
```

```python
# Keep only English-language movies
movies_cleaned = movies_cleaned[movies_cleaned['original_language'] == 'en']

print(movies_cleaned['original_language'].value_counts())
```

```python
movies_cleaned['release_date'] = pd.to_datetime(movies_cleaned['release_date'], errors='coerce')

# Quick check
print(movies_cleaned['release_date'].dtype)
print(movies_cleaned.info())
movies_cleaned.head()
```

```python
movies_cleaned['release_year'] = movies_cleaned['release_date'].dt.year
movies_cleaned['release_month'] = movies_cleaned['release_date'].dt.month
movies_cleaned.info()
```

```python
# Drop popularity, vote_average, vote_count columns
cols_to_drop = ['popularity', 'vote_average', 'vote_count']
movies_cleaned = movies_cleaned.drop(columns=cols_to_drop)

movies_cleaned.head()
```

```python
#Check status values
print(movies_cleaned['status'].value_counts())
```

```python
# Drop original_language and status columns -- All titles are english and released
movies_cleaned = movies_cleaned.drop(columns=['original_language', 'status'])
movies_cleaned.head()
```

```python
#Reorder Columns

cols = list(movies_cleaned.columns)
cols.remove('release_date')
cols.insert(-2, 'release_date')

# Reorder dataframe
movies_cleaned = movies_cleaned[cols]

movies_cleaned.head()
```

```python
genre_counts = movies_cleaned['genres'].value_counts()
print(genre_counts.head(200))
```

```python
# Keep only the first two genres
movies_cleaned['genres'] = movies_cleaned['genres'].apply(lambda x: '-'.join(x.split('-')[:2]))

# check
movies_cleaned['genres'].value_counts()

top_genres = movies_cleaned['genres'].value_counts().head(20)
print(top_genres)
```

```python
# 5 movies with 'comedy-drama'
comedy_drama_movies = movies_cleaned[movies_cleaned['genres'] == 'Comedy-Drama'].head(5)
print("Comedy-Drama Movies:")
print(comedy_drama_movies[['title', 'genres', 'release_year']])

# 5 movies with 'drama-comedy'
drama_comedy_movies = movies_cleaned[movies_cleaned['genres'] == 'Drama-Comedy'].head(5)
print("\nDrama-Comedy Movies:")
print(drama_comedy_movies[['title', 'genres', 'release_year']])
```

```python
# Create a column with the first two genres
movies_cleaned['top_2_genres'] = movies_cleaned['genres'].apply(lambda x: '-'.join(x.split('-')[:2]))

# Reorder columns to place it next to 'genres'
cols = list(movies_cleaned.columns)
genre_index = cols.index('genres')
# Remove 'top_2_genres' from its current position if it exists
cols.remove('top_2_genres')
# Insert it right after 'genres'
cols.insert(genre_index + 1, 'top_2_genres')

# Apply the new column order
movies_cleaned = movies_cleaned[cols]

# check
movies_cleaned[['genres', 'top_2_genres']].head(10)
```

```python
# Keep only the first genre
movies_cleaned['genres'] = movies_cleaned['genres'].apply(lambda x: x.split('-')[0])

movies_cleaned = movies_cleaned.rename(columns={'genres': 'genre'})

# Quick check
movies_cleaned['genre'].value_counts()
```

```python
movies_cleaned.head(30)
```

```python
# Extract the first production company
movies_cleaned['top_production_company'] = movies_cleaned['production_companies'].apply(lambda x: x.split('-')[0].strip())

# check
movies_cleaned['top_production_company'].value_counts()
```

```python
movies_cleaned.head(30)
```

```python
# Drop the original production_companies column
movies_cleaned = movies_cleaned.drop(columns=['production_companies'])

# Quick check
movies_cleaned.head()
```

```python
# Get current columns
cols = list(movies_cleaned.columns)

# Remove 'top_production_company' from current position
cols.remove('top_production_company')

# Insert at index 4 (5th position)
cols.insert(4, 'top_production_company')

# Reorder dataframe
movies_cleaned = movies_cleaned[cols]

# check
movies_cleaned.head()
```

```python
# Replace 'Metro' with 'Metro-Goldwyn-Mayer'
movies_cleaned['top_production_company'] = movies_cleaned['top_production_company'].replace({'Metro': 'Metro-Goldwyn-Mayer'})

# check
movies_cleaned['top_production_company'].value_counts().head(30)
```

```python
# Split the credits column into up to 4 actors
actors_split = movies_cleaned['credits'].str.split('-', expand=True)

# Assign to new columns
movies_cleaned['actor_1'] = actors_split[0].str.strip()
movies_cleaned['actor_2'] = actors_split[1].str.strip()
movies_cleaned['actor_3'] = actors_split[2].str.strip()
movies_cleaned['actor_4'] = actors_split[3].str.strip() 

# Drop the original credits column
movies_cleaned = movies_cleaned.drop(columns=['credits'], errors='ignore')

# Quick check
movies_cleaned[['actor_1', 'actor_2', 'actor_3', 'actor_4']].head(10)

movies_cleaned.head()


```

```python
# Count rows before cleaning
before_count = len(movies_cleaned)

# Remove rows where budget or revenue is 0 or missing
movies_cleaned = movies_cleaned[
    (movies_cleaned['budget'] > 0) & (movies_cleaned['revenue'] > 0)
].copy()

# Reset index for cleanliness
movies_cleaned.reset_index(drop=True, inplace=True)

# Count rows after cleaning
after_count = len(movies_cleaned)

# Calculate how many rows were removed
removed_count = before_count - after_count

# Display summary
print({before_count})
print({after_count})
print({removed_count})
```

```python

# Calculate ROI as a whole-number percentage
movies_cleaned['return_on_investment'] = (
    (movies_cleaned['revenue'] - movies_cleaned['budget']) / movies_cleaned['budget'] * 100
).round(0)

# Move ROI column to 7th position (index 6)
roi_col = movies_cleaned.pop('return_on_investment')
movies_cleaned.insert(7, 'return_on_investment', roi_col)

movies_cleaned.head(20)
```

```python
# Calculate profit
movies_cleaned['profit'] = movies_cleaned['revenue'] - movies_cleaned['budget']

profit_col = movies_cleaned.pop('profit')
movies_cleaned.insert(7, 'profit', profit_col)  
```

```python
movies_cleaned = movies_cleaned.drop(columns=['runtime'], errors='ignore')

movies_cleaned.head()
```

```python
print(f"Shape after cleaning: {movies_cleaned.shape}")
print(movies_cleaned.isnull().sum())
```

```python
# Filter movies where at least one of actor_2, actor_3, or actor_4 is null
movies_some_missing_actors = movies_cleaned[
    movies_cleaned[['actor_2', 'actor_3', 'actor_4']].isna().any(axis=1)
]

# Display relevant columns
movies_some_missing_actors[['title', 'actor_1', 'actor_2', 'actor_3', 'actor_4']]
```

```python
# Fill missing actor columns with 'NA'
movies_cleaned[['actor_2', 'actor_3', 'actor_4']] = movies_cleaned[['actor_2', 'actor_3', 'actor_4']].fillna('NA')

# check
print(f"Shape after cleaning: {movies_cleaned.shape}")
print(movies_cleaned.isnull().sum())
```
